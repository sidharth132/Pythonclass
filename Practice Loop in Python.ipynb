```python
year = int(input("Enter a year: "))  
if (year % 4) == 0:  
   if (year % 100) == 0:  
       if (year % 400) == 0:  
           print("{0} is a leap year".format(year))  
       else:  
           print("{0} is not a leap year".format(year))  
   else:  
       print("{0} is a leap year".format(year))  
else:  
   print("{0} is not a leap year".format(year))  
```

```python
n=int(input("Enter any number"))
row=0
while row<n:
    star=row+1
    while star>0:
        print("*",end=" ")
        star=star-1
    row=row+1
    print("\r")
```
